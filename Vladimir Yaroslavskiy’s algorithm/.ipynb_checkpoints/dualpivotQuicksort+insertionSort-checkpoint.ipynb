```python
import numpy as np
import time

# Python3 program to implement  

# Function to do insertion sort
def insertionSort(arr):
 
    # Traverse through 1 to len(arr)
    for i in range(1, len(arr)):
 
        key = arr[i]
 
        # Move elements of arr[0..i-1], that are
        # greater than key, to one position ahead
        # of their current position
        j = i-1
        while j >= 0 and key < arr[j] :
                arr[j + 1] = arr[j]
                j -= 1
        arr[j + 1] = key

# dual pivot QuickSort 
def vladimirSort(arr, low, high, L): 
    if low < high: 
        if (len(arr) < L): #L = 17, 33, 65, 129, 257, 513
            insertionSort(arr)
            return
          
        # lp means left pivot and rp  
        # means right pivot 
        lp, rp = partition(arr, low, high) 
          
        vladimirSort(arr, low, lp - 1, L) 
        vladimirSort(arr, lp + 1, rp - 1, L) 
        vladimirSort(arr, rp + 1, high, L) 
        
def partition(arr, low, high): 
      
    if arr[low] > arr[high]: 
        arr[low], arr[high] = arr[high], arr[low] 
          
    # p is the left pivot, and q is the right pivot. 
    j = k = low + 1
    g, p, q = high - 1, arr[low], arr[high] 
      
    while k <= g: 
          
        # If elements are less than the left pivot 
        if arr[k] < p: 
            arr[k], arr[j] = arr[j], arr[k] 
            j += 1
              
        # If elements are greater than or equal  
        # to the right pivot 
        elif arr[k] >= q: 
            while arr[g] > q and k < g: 
                g -= 1
                  
            arr[k], arr[g] = arr[g], arr[k] 
            g -= 1
              
            if arr[k] < p: 
                arr[k], arr[j] = arr[j], arr[k] 
                j += 1
                  
        k += 1
          
    j -= 1
    g += 1
      
    # Bring pivots to their appropriate positions. 
    arr[low], arr[j] = arr[j], arr[low] 
    arr[high], arr[g] = arr[g], arr[high] 
      
    # Returning the indices of the pivots 
    return j, g 


# Driver code 

arr= np.random.randint(1,101,10000) # array size can be 10k, 100k, 200k, 400k, 800k, 1600k, 3200k, 6400k
  
start = time.time()  #수행시간 측정 시작

vladimirSort(arr, 0, len(arr)-1, 17) #L = 17, 33, 65, 129, 257, 513

end = time.time() # 수행시간 측정 완료

print(end - start)
```
