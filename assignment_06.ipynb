## 유클리드 호제법 사용하기
유클리드 호제법은 x, y 의 최대공약수는 y, r(r은 x와 y의 나머지)의 최대공약수와 같다는 원리를 이용하는 것입니다.

즉, 계속해서 x 값에는 y 값을 대입하고 y 값에는 r 값을 대입하다보면 언젠가는 r 이 0 이 되는데 이때에 y 값에 있는 값이 최대공약수 입니다.



## 최소공배수 공식
: 두 자연수의 곱 / 최대공약수

```python
a, b = map(int, input().split())

multiple = a*b
while(a % b !=0):
    r = a%b
    a = b
    b = r
    
gcd = b

lcd = multiple // gcd
        
print(gcd)
print(lcd)
```
